# Get avg for each category and overall avg

```python
def get_cat_col_names(all_cols, cat):
    # include the target == cat but not source == cat
    col_targets = [(cat in x.split("to_")[1] and cat not in x.split("to_")[0]) for x in all_cols if x!="mean"]
    col_targets.append(False)
    return np.array(all_cols)[col_targets]
```

```python
def get_cat_avgs(df, cat):
    cat_cols = get_cat_col_names(df.columns, cat)
    return {cat:df[cat_cols].mean(axis=1)}
```

```python
def get_all_cat_avg(df):
    final_dict = {}
    for x in cat_mapping_dict.values():
        final_dict.update(get_cat_avgs(df, x))
                          
    final_dict["mean"] = df['mean'].values
    return final_dict
```

```python
with pd.ExcelWriter('../results.xlsx') as writer:  
    pd.DataFrame.from_dict(get_all_cat_avg(no_fs_combo)).to_excel(writer, sheet_name='no_fs_avg')
    pd.DataFrame.from_dict(get_all_cat_avg(fs_least_df)).to_excel(writer, sheet_name='fs_least_avg')
    pd.DataFrame.from_dict(get_all_cat_avg(fs_random_df)).to_excel(writer, sheet_name='fs_random_avg')
    no_fs_combo.to_excel(writer, sheet_name='no_fs')
    fs_least_df.to_excel(writer, sheet_name='fs_least')
    fs_random_df.to_excel(writer, sheet_name='fs_random')
```
